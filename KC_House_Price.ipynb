```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))
# optimizer = 'adam' - regression model
model.compile(optimizer='adam', loss='mse')
```

```python
# validation_data is used to check the error, but does not change the weights of the model
model.fit(x = X_train, y = y_train,
          validation_data=(X_test, y_test),
          batch_size=128, epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
# the graph denotes that the loss vs val_loss is similar and reducing as the epochs move higher. The model can be trained further without overfitting.
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```
